```python
import json
from config import filename

with open(filename) as f:
    data = json.load(f)
```

```python
from typing import List

def summary(l: List[int]):
    
    if len(l) == 0:
        return 0, 0, 0, 0
    
    s = sum(l)
    m = s / len(l)    
    total_var = sum((x - m) ** 2 for x in l)
    var = total_var / len(l)
    
    return s, m, total_var, var
```

```python
from collections import defaultdict

eventDuration = defaultdict(lambda : [])

for event in data['traceEvents']:
    
    if 'name' not in event:
        continue
        
    if 'args' not in event:
        continue
        
    if 'dur' not in event:
        continue
        
    if 'Input Dims' not in event['args']:
        continue
        
    concrete_name = f'{event["name"]} | {event["ph"]} | {event["args"]["Input Dims"]}'
    
    eventDuration[concrete_name].append(event['dur'])
        
summary_all = {k: summary(v) for k, v in eventDuration.items()}
```
